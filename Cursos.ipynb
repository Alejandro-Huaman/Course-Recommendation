```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import openai

#openai.api_key = "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"

#prompt = "Escribe aquí el inicio del texto que deseas generar..."

# Definir una lista para almacenar todos los cursos
todos_los_cursos = []
cursos_solo_titulos = []

# Iterar a través de todas las páginas (desde la página 1 hasta la página 25)
for pagina in range(1, 26):
    url = f'https://campusromero.pe/cursos/page/{pagina}/'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        cursos = soup.find_all('div', class_='box-date')
        
        for curso in cursos:
            titulo = curso.find('h3').text.strip()
            url = curso.find_next('a')['href'] 

            course_response = requests.get(url)
            course_soup = BeautifulSoup(course_response.text, 'html.parser')
            
            course_description_box = course_soup.find('div', class_='contenido-standard')
            
            course_description_element = course_description_box.find('p')
            if course_description_element is not None:
                course_description = course_description_element.text.strip()
            else:
                course_description = course_description_box.find('font').text.strip()             

            todos_los_cursos.append({
                'Titulo': titulo,
                'URL': url,
                'Descripción': course_description,
            })

            cursos_solo_titulos.append(titulo)

# Imprimir los cursos obtenidos
for i, curso in enumerate(todos_los_cursos, start=1):
    print(f'Curso {i}:')
    print(f"Título: {curso['Titulo']}")
    print(f"Descripción: {curso['Descripción']}")
    print(f"Url: {curso['URL']}")
    print('--------------------')

print(cursos_solo_titulos)

#OpenAI GPT para la recomendacion de cursos
```
